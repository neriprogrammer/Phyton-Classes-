```python
import yfinance as yf
```

```python
tickers = ['PETR4.SA', 'VALE3.SA', 'WEGE3.SA','BPAC11.SA','RADL3.SA', 'OIBR3.SA','BOVA11.SA']
```

```python
inicio = '2020-01-01'
fim = '2022-02-28'
```

```python
precos = yf.download(tickers, 
                     start = inicio, 
                     end = fim)['Adj Close']
```

```python
normalizado = precos/precos.iloc[0]
```

```python
normalizado.iloc[[0, -1]]. transpose()*100
```

```python
normalizado.plot(figsize = (10, 10));
```
